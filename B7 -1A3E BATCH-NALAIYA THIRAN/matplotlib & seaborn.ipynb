## Histogram

```python
x2=[1,1,2,4,5,3,5,6,7,9,8,7,5,6,6,6,4,3,3,8,9]
```

```python
plt.hist(x2)
```

```python
plt.hist(x2)
plt.xlabel('x2 values')
plt.ylabel('freq.of x2 values')
plt.title('histogram of x2')
plt.show()
```
