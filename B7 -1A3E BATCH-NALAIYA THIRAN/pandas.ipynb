## merge

```python
ds=pd.DataFrame({'name':['Anu','Appu','Ammu','Arun'],'History':[89,97,88,79]})
ds
```

```python
dq=pd.DataFrame({'name':['Anu','Appu','Ammu','Arun'],'Tamil':[92,95,90,93]})
dq
```

```python
dw=pd.DataFrame({'name':['Anu','Appu','Ammu','Arun'],'English':[93,94,91,96]})
dw
```

```python
m1=pd.merge(ds,dq,on='name')
m1
```

```python
final_merge=pd.merge(m1,dw,on='name')
final_merge
```
